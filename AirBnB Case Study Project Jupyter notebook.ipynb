```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head()
```

```python

data = pd.DataFrame.from_dict(air_bnb)

data
```

```python
air_bnb = pd.read_csv('./AB_NYC_2019.csv', sep=',')
```

```python
bus_model = air_bnb.query('availability_365 == 0')
control = bus_model.drop(columns=['id','neighbourhood_group',
       'neighbourhood', 'latitude', 'longitude', 'room_type', 'price',
       'minimum_nights', 'number_of_reviews', 'last_review',
       'reviews_per_month','name', 'availability_365'])
control
```

```python
#keys 
print(type(control))
```

```python
count_data = data.sort_values('calculated_host_listings_count', kind='mergesort', ascending=False)

count_data.query('calculated_host_listings_count >100')
```

```python
#Which hosts are the busiest and why?
air_bnb.groupby(['host_id','host_name']).sum().sort_values('calculated_host_listings_count', ascending=False)['calculated_host_listings_count']
```

```python
# How many neighborhood groups are available and which shows up the most?
air_bnb.groupby('neighbourhood_group').count()['id'].sort_values(ascending=False)
```

```python
air_bnb['neighbourhood_group'].value_counts()
```

```python
# Are private rooms the most popular in manhattan?
air_bnb.groupby(['neighbourhood_group', 'room_type'])['id'].count()
```

```python
# Which hosts are the busiest and based on their reviews?
air_bnb.groupby(['host_id', 'host_name']).sum().number_of_reviews.nlargest(10)
```

```python
air_bnb.groupby(['host_id', 'host_name']).sum().reviews_per_month.nlargest(10)
```

```python
#Which neighorhood group has the highest average price?
air_bnb.groupby('neighbourhood_group').price.mean().sort_values(ascending=False)
```

```python
# Which neighbor hood group has the highest total price?
air_bnb.groupby('neighbourhood_group').price.max().sort_values(ascending=False)
```

```python
air_bnb.groupby(['host_id', 'host_name']).price.sum().sort_values(ascending=False).nlargest(5)
```

```python
#Which top 5 hosts have the highest total price?
air_bnb.groupby(['host_id', 'host_name']).price.sum().sort_values(ascending=False).nlargest(5)
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
air_bnb[(air_bnb['availability_365']==0) & (air_bnb['number_of_reviews'] >= 100)]
```

```python
air_bnb.query('availability_365 == 0 and number_of_reviews >= 100')
```

```python
# What host has the highest total of prices and where are they located?
air_bnb.groupby(['host_name', 'neighbourhood']).price.sum().nlargest(5)
```

```python
# When did Danielle from Queens last receive a review?
air_bnb.query("host_name == 'Danielle' and neighbourhood_group == 'Queens'").last_review.sort_values(ascending=False)
```

```python
air_bnb.query("host_name == 'Danielle' and neighbourhood_group == 'Queens'").groupby('host_id').count()
```

```python
air_bnb.query("host_name == 'Danielle' and neighbourhood_group == 'Queens'").groupby(['host_id', 'host_name']).last_review.max()
```
